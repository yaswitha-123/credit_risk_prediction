```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances_argmin_min
import numpy as np

kmeans = KMeans(n_clusters=2, random_state=42)
dbscan = DBSCAN(eps=0.5, min_samples=5)
agg_clustering = AgglomerativeClustering(n_clusters=2)

kmeans_labels = kmeans.fit_predict(df)
dbscan_labels = dbscan.fit_predict(df)
agg_labels = agg_clustering.fit_predict(df)

dbscan_labels = np.where(dbscan_labels == -1, 0, dbscan_labels) 
labels_matrix = np.vstack([kmeans_labels, dbscan_labels, agg_labels]).T

from scipy.stats import mode
ensemble_labels, _ = mode(labels_matrix, axis=1)

df['Ensemble_Cluster'] = ensemble_labels.flatten()

df['Risk'] = df['Ensemble_Cluster'].apply(lambda x: 'Risk' if x == 1 else 'Non-Risk')

X = df.drop(columns=['Ensemble_Cluster', 'Risk'])
y = df['Risk']
```
